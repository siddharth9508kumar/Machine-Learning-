```python
#1 knn implementation to predict the class for a new data point.
import numpy as np
import math
x=np.array([[2,4],[4,2],[4,4],[6,2],[6,4]])
y=np.array(['A','A','B','B','B'])
# to implement knn
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x,y)
knn.predict([[5,3]])
print(knn.predict([[5,3]]))


#function to calculate euclidean distance
def euclidean_distance(x1,x2):
    return math.sqrt(sum((x1-x2)**2 for x1,x2 in zip(x1,x2)))
print(euclidean_distance(x[0],x[1]))



```

```python
#Q2 download dataset from kaggle and implement knn and evaluate it..
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
data=pd.read_csv('diabetes.csv')
data.head()
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(x,y)
knn.predict([[1,85,66,29,0,26.6,0.351,31]])
print("knn value is",knn.predict([[1,85,66,29,0,26.6,0.351,31]]))
#To evaluate it
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
y_pred=knn.predict(x)
print("accuracy:",accuracy_score(y,y_pred))
print("confusion matrix:",confusion_matrix(y,y_pred))
print("classification report:",classification_report(y,y_pred))

```

```python
#Q3 to implemment naive bayes classifier
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
data=pd.read_csv('diabetes.csv')
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
gnb=GaussianNB()
gnb.fit(x,y)
y_pred=gnb.predict(x)
print(y_pred[:5])
#to predict new value..
print(gnb.predict([[6,148,72,35,0,33.6,0.627,50]]))

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y,y_pred))
print(confusion_matrix(y,y_pred))
print(classification_report(y,y_pred))


```
